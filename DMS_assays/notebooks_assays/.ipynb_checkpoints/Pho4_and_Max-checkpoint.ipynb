```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
```

# Import Pho4 data

```python
dna_sequences = ['CAACGTGA_CAGCGTGA', 'CCACGCGA_CCACGTGA', 'CCACGTTA_CCACGTGA', 'CCATGTGA_CCACGTGA', 'TCACGTGC_GCACGTGC']

dfs = {}
for dna_seq in dna_sequences:
    csv_filename = f"../raw_data/PHO4_YEAST_Aditham_2021/200328_PHO4_{dna_seq}.csv"
    df = pd.read_csv(csv_filename, index_col=0)

    col1 = f'ddG_wt_{dna_seq.split('_')[0]}
    col2 = f'estimate_{dna_seq.split('_')[0]}'
    
    df = df[['id', col1, col2]].drop_duplicates()

    # Assuming your dataframe is called `df`
    # Group by 'id' and aggregate mean and SEM
    summary_df = df.groupby('id').agg({
        col1: ['mean', sem],
        col2: ['mean', sem]
    })
    
    # Flatten the MultiIndex columns
    summary_df.columns = [f'{col1}_mean', f'{col1}_sem', f'{col2}_mean', f'{col2}_sem']
```
