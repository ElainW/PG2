```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
```

```python
raw_data = pd.read_csv("../raw_data/MNTH_DEIRA_Berry_2025/merged_Mn_Mg_scores_all.csv", index_col=0)
raw_data = raw_data.set_index(pd.Index([i.replace(',','_') for i in raw_data.index]))
raw_data['mutant'] = raw_data.index
```

```python
raw_data[['aa_seq', 'Mn_import_score', 'Mn_import_score_err']].to_csv("../processed_data/MNTH_DEIRA_Berry_2025/MNTH_DEIRA_Berry_2025_Mn.csv")
raw_data[['aa_seq', 'Mg_import_score', 'Mg_import_score_err']].to_csv("../processed_data/MNTH_DEIRA_Berry_2025/MNTH_DEIRA_Berry_2025_Mg.csv")
```

```python
raw_data[['aa_seq', 'Mn_import_score', 'Mn_import_score_err', 'mutant']].to_csv("~/lab/SpecificityStudio/DMS_assays/MNTH_DEIRA_Berry_2025_Mn.csv")
raw_data[['aa_seq', 'Mg_import_score', 'Mg_import_score_err', 'mutant']].to_csv("~/lab/SpecificityStudio/DMS_assays/MNTH_DEIRA_Berry_2025_Mg.csv")
```
