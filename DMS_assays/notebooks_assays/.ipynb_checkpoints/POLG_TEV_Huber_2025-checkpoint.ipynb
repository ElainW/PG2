```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv("../raw_data/POLG_TEV_Huber_2025/POLG_TEV_Huber_Kucera_2025.csv")
```

```python
wt_seq = ['' for i in range(500)]
for i in raw_data['TEVp']:
    if '_' not in i:
        res = i[0]
        try:
            pos = int(i[1:-1])
        except:
            None
        wt_seq[pos]=res
        if pos<2:
            print(i)
```
