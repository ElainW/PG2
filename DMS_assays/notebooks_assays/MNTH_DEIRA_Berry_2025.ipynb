```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
```

```python
raw_data = pd.read_csv("../raw_data/MNTH_DEIRA_Berry_2025/merged_Mn_Mg_scores_all.csv", index_col=0)
raw_data = raw_data.set_index(pd.Index([i.replace(',',':') for i in raw_data.index]))

raw_data = raw_data.loc[['del' not in i and 'ins' not in i and 'no_start' not in i for i in raw_data.index]]
raw_data = raw_data.rename({"WT":"M1M"})

raw_data['mutant'] = raw_data.index
raw_data['mutated_sequence'] = raw_data['aa_seq']

```

```python
raw_data[['aa_seq', 'Mn_import_score', 'Mn_import_score_err', 'mutated_sequence']].to_csv("../processed_data/MNTH_DEIRA_Berry_2025/MNTH_DEIRA_Berry_2025_Mn.csv")
raw_data[['aa_seq', 'Mg_import_score', 'Mg_import_score_err', 'mutated_sequence']].to_csv("../processed_data/MNTH_DEIRA_Berry_2025/MNTH_DEIRA_Berry_2025_Mg.csv")
```
